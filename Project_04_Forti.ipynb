## Part 3: Working with Categorical Variables
Create lists to specify the order for each of the three categorical variables.

```python
# list of categorical variable's levels from worst -> best
clarity_levels = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']
cut_levels = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
color_levels = ['J', 'I', 'H', 'G', 'F', 'E', 'D']
```

Use the lists to communicate to Pandas the correct order for the levels of the three categorical variables.

```python
# convert to categorical variables
diamonds.clarity = pd.Categorical(diamonds.clarity, clarity_levels)
diamonds.cut = pd.Categorical(diamonds.cut, cut_levels)
diamonds.color = pd.Categorical(diamonds.color, color_levels)
```

Create lists of named colors to serve as palettes to be used for visualizations later.

```python
# lists of colors
clarity_pal = ['firebrick', 'deeppink', 'coral', 'gold', 'limegreen', 'aqua', 'blue', 'darkviolet']
cut_pal = ['deeppink', 'gold', 'limegreen', 'aqua', 'darkviolet']
color_pal = ['firebrick', 'deeppink', 'coral', 'gold', 'limegreen', 'aqua', 'darkviolet']
```

## Part 5: Scatterplots of Price Against Carat
Create scatter plots to explore the relationship between **price** and **carat**.

```python
plt.figure(figsize=[8,6])

# make scatter plot of price vs carat
plt.scatter(x=diamonds.loc[:, 'carat'],
            y=diamonds.loc[:, 'price'],
            s=20,
            alpha=0.4,
            color='hotpink')

# add labels
plt.xlabel('Carat Size')
plt.ylabel('Price (in $)')
plt.title('Relationship between Diamond Price and Carat Size')

plt.show()
```

Reproduce the plot from above, but will now color the points according to their clarity.

```python
plt.figure(figsize=[8,6])

# make scatter plot of price vs carat
for level in range(len(clarity_levels)):
    plt.scatter(x=diamonds.loc[diamonds.loc[:, 'clarity']==clarity_levels[level], 'carat'],
                y=diamonds.loc[diamonds.loc[:, 'clarity']==clarity_levels[level], 'price'],
                s=20,
                alpha=0.4,
                color=clarity_pal[level])

# add labels and legend
plt.legend(clarity_levels)
plt.xlabel('Carat Size')
plt.ylabel('Price (in $)')
plt.title('Relationship between Diamond Price and Carat Size')

plt.show()
```

Separate the points for each of the eight clarity levels into its own subplot.

```python
plt.figure(figsize=[12,6])

# make scatter plot of price vs carat
for level in range(len(clarity_levels)):
    plt.subplot(2,4,level+1)
    plt.scatter(x=diamonds.loc[diamonds.loc[:, 'clarity']==clarity_levels[level], 'carat'],
                y=diamonds.loc[diamonds.loc[:, 'clarity']==clarity_levels[level], 'price'],
                s=20,
                alpha=0.4,
                color=clarity_pal[level])

    # add axes limits
    plt.xlim([0,4])
    plt.ylim([0,20000])
    
    # add labels
    plt.xlabel('Carat Size')
    plt.ylabel('Price (in $)')
    plt.title(clarity_levels[level])

plt.tight_layout()
plt.show()
```

## Part 6: Applying Logarithmic Transformations
Apply logarithmic transformations to **price** and **carat**, and explore the distribution of these transformed variables. Start by adding two new columns to the **diamonds** DataFrame to store the transformed variables.

```python
# add cols for ln of carat and price
diamonds['ln_carat'] = np.log(diamonds.loc[:,'carat'])
diamonds['ln_price'] = np.log(diamonds.loc[:,'price'])
```

Explore the distribution of the diamond prices and the log of the diamond prices with histograms.

```python
plt.figure(figsize=[12,4])

# make histogram of price
plt.subplot(1,2,1)
plt.hist(x=diamonds.loc[:, 'price'],
         edgecolor='k',
         color='dodgerblue',
         bins=20)
# add labels
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Histogram of Prices')

# make histogram of ln_price
plt.subplot(1,2,2)
plt.hist(x=diamonds.loc[:, 'ln_price'],
         edgecolor='k',
         color='dodgerblue',
         bins=20)
# add labels
plt.xlabel('Natural Log of Price')
plt.ylabel('Count')
plt.title('Histogram of Log-Prices')

plt.show()
```

Explore the distribution of the carat sizes and the log of the carat sizes with histograms.

```python
plt.figure(figsize=[12,4])

# make histogram of carat
plt.subplot(1,2,1)
plt.hist(x=diamonds.loc[:, 'carat'],
         edgecolor='k',
         color='limegreen',
         bins=20)
# add labels
plt.xlabel('Carat Size')
plt.ylabel('Count')
plt.title('Histogram of Carat Sizes')

# make histogram of ln_carat
plt.subplot(1,2,2)
plt.hist(x=diamonds.loc[:, 'ln_carat'],
         edgecolor='k',
         color='limegreen',
         bins=20)
# add labels
plt.xlabel('Natural Log of Carat Size')
plt.ylabel('Count')
plt.title('Histogram of Log-Carat Sizes')

plt.show()
```

## Part 7: Scatterplot of Transformed Variables
Create a scatterplot of **ln_price** against **ln_carat**, with points colored according to their **clarity**.

```python
plt.figure(figsize=[8,6])

# make scatter plot of ln_price vs ln_carat
for level in range(len(clarity_levels)):
    plt.scatter(x=diamonds.loc[diamonds.loc[:, 'clarity']==clarity_levels[level], 'ln_carat'],
                y=diamonds.loc[diamonds.loc[:, 'clarity']==clarity_levels[level], 'ln_price'],
                s=20,
                alpha=0.4,
                color=clarity_pal[level])
# add labels and legend
plt.legend(clarity_levels)
plt.xlabel('Natural Log of Carat Size')
plt.ylabel('Natural Log of Price')
plt.title('Relationship between Log-Price and Log-Carat Size')

plt.show()
```

## Part 8: Grouping by Cut
Create a grouped DataFrame displaying the mean price and carat size for each cut level.

```python
# group selected cols by cut and calc mean
gb_cut = diamonds.loc[:,['cut', 'price', 'carat']].groupby('cut').mean()
display(gb_cut)
```

Graphically display the information from the DataFrame with a bar chart.

```python
plt.figure(figsize = [12,4])

# plot bar chart for price
plt.subplot(1,2,1)
plt.bar(x=gb_cut.index.values,
        height=gb_cut.loc[:, 'price'],
        color=cut_pal,
        edgecolor='k')
# set title and labels
plt.title('Mean Price by Cut')
plt.xlabel('Cut')
plt.ylabel('Mean Price')

# plot bar chart for carat
plt.subplot(1,2,2)
plt.bar(x=gb_cut.index.values,
        height=gb_cut.loc[:, 'carat'],
        color=cut_pal,
        edgecolor='k')
# set title and labels
plt.title('Mean Carat Size by Cut')
plt.xlabel('Cut')
plt.ylabel('Mean Carat Size')

plt.show()
```
